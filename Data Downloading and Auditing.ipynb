```python
# Convert date columns to datetime format
#sales_df["order date (DateOrders)"] = pd.to_datetime(sales_df["order date (DateOrders)"], errors='coerce')
#sales_df["shipping date (DateOrders)"] = pd.to_datetime(sales_df["shipping date (DateOrders)"], errors='coerce')

sales_df["order_date"] = pd.to_datetime(sales_df["order date (DateOrders)"], errors="coerce", format="mixed")
sales_df["shipping_date"] = pd.to_datetime(sales_df["shipping date (DateOrders)"], errors="coerce", format="mixed")
```

```python
# Ensure order date is always before shipping date
sales_df[sales_df["order_date"] > sales_df["shipping_date"]]
```

```python
# Check for Inconsistent or Invalid Values: Negative sales, order quantities, or profits.
sales_df[sales_df["Sales"] < 0]
```

```python
print(sales_df["Order Id"].apply(lambda x: str(x).isalnum()).value_counts())
```

```python
### Adding an aditional column

sales_df['late_delivery_risk'] = sales_df.apply(
    lambda x: 'Late' if x['Days for shipping (real)'] > x['Days for shipment (scheduled)'] else 'Not Late',
    axis=1
)

# Rename columns using Python naming conventions
sales_df.rename(columns=lambda x: x.strip().lower().replace(" ", "_").replace("(", "").replace(")", ""), inplace=True)

# Display first few rows to verify changes
sales_df.head()
```

```python
# Adding one more column : full name of customer
sales_df["customer_name"] = sales_df["customer_fname"] + " " + sales_df["customer_lname"].fillna("")
sales_df["customer_name"] = sales_df["customer_name"].str.strip()  # Remove any extra spaces
```

```python
# Droping unwanted columns now
cols_to_drop = ['customer_fname', 'customer_lname','order_date_dateorders','shipping_date_dateorders']

sales_df = sales_df.drop(columns=cols_to_drop)
```

## Loading data into Postgresql

```python
# Connection Details
from sqlalchemy import create_engine

# Define PostgreSQL connection parameters
db_name = "inventory_management"  # Change this if using a new DB
db_user = "postgres"
db_password = "Royal%402025"
db_host = "localhost"  # Change if hosted remotely
db_port = "5432"  # Default PostgreSQL port

# Create database connection
engine = create_engine(f"postgresql+psycopg2://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}")
```
